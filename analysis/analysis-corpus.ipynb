```python
import pandas as pd
import json

with open('../data/corpus.json') as f:
    data = json.load(f)
    df = pd.json_normalize(data)
    df.head()

```

```python
df.head()

def pattern_length(row):
    
    return len(str(row['pattern']))


df['pattern_length'] = df.apply (lambda row: pattern_length(row), axis=1)

df['pattern_length'].plot(kind="box", ylabel='Pattern length')
```

```python
df['pattern_length'].describe()
```

```python
from matplotlib import pyplot as plt


binned = pd.cut(df['pattern_length'], bins=[0,10,30, 50, 100, 500, 1000, 5000, 10000, 162866])

ax = binned.value_counts(sort=False).plot.bar(rot=90, color="b", figsize=(6,4))
ax.set_xlabel("Pattern length")
ax.set_ylabel("Count")
plt.bar_label(ax.containers[0])

plt.legend()
plt.show()
```
